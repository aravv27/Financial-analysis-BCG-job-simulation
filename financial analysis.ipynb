```python
import pandas as pd
import numpy as np
```

```python
msc = pd.read_csv(r"D:\FIN GEN-AI\DATA\MICROSOFT.csv")
app = pd.read_csv(r"D:\FIN GEN-AI\DATA\APPLE.csv")
tes = pd.read_csv(r"D:\FIN GEN-AI\DATA\TESLA.csv")

for col in msc.columns[1:]:
    msc[col] = msc[col].replace(",", "", regex=True).astype(float)
for col in app.columns[1:]:
    app[col] = app[col].replace(",", "", regex=True).astype(float)
for col in tes.columns[1:]:
    tes[col] = tes[col].replace(",", "", regex=True).astype(float)
msc.head()
```

```python
msc.rename(columns={'Unnamed: 0': 'Metric'}, inplace=True)
app.rename(columns={'Unnamed: 0': 'Metric'}, inplace=True)
tes.rename(columns={'Unnamed: 0': 'Metric'}, inplace=True)

print(tes.head())
```

```python
df_msc = msc.melt(id_vars=['Metric'], var_name='Year', value_name='Value')
df_app = app.melt(id_vars=['Metric'], var_name='Year', value_name='Value')
df_tes = tes.melt(id_vars=['Metric'], var_name='Year', value_name='Value')

df_msc['Company'] = 'MSC'
df_app['Company'] = 'APP'
df_tes['Company'] = 'TES'

df_msc = df_msc.pivot_table(index=['Company', 'Year'], columns='Metric', values='Value').reset_index()
df_app = df_app.pivot_table(index=['Company', 'Year'], columns='Metric', values='Value').reset_index()
df_tes = df_tes.pivot_table(index=['Company', 'Year'], columns='Metric', values='Value').reset_index()

df = pd.concat([df_msc, df_app, df_tes], ignore_index=True)
```

```python
df.head(10)
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['TOTAL REVENUE'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['NET INCOME'].pct_change() * 100
df['CashFlow Growth (%)'] = df.groupby(['Company'])['NET CASH FROM OPERATIONS'].pct_change()*100
```

```python
df.head()
```

```python
df['Revenue Growth (%)'].fillna(0, inplace=True)
df['Net Income Growth (%)'].fillna(0, inplace=True)
df['CashFlow Growth (%)'].fillna(0,inplace = True)
```

```python
df.head()
```

```python
df['Debt Ratio (%)'] = (df['TOTAL LIABILITIES'] / df['TOTAL ASSETS'])*100
```

```python
df["Asset Turnover (%)"] = (df["TOTAL REVENUE"]/df["TOTAL ASSETS"])*100
```

```python
df["ROA (%)"] = (df["NET INCOME"]/df["TOTAL ASSETS"])*100
```

```python
df["Cash ROA(%)"] = (df["NET CASH FROM OPERATIONS"]/df["TOTAL ASSETS"])*100
df.head()
```

```python
df.describe()
```

```python
df_melted = pd.melt(df, id_vars=["Company", "Year"], var_name="Metric", value_name="Value")
#features = df["Metric"].unique()
df_melted.head()
#print(features)
```

```python
features = df_melted["Metric"].unique().tolist()
print(features)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "whitegrid")
# features = (df.columns).tolist()
# features.remove("Company")
# features.remove("Year")
# print(features)
# print(len(features))
```

```python
fig, axs = plt.subplots(4, 3, figsize=(18, 20))
axs = axs.flatten()


for i, feature in enumerate(features):
    feature_data = df_melted[df_melted["Metric"] == feature]
    
    sns.barplot(data=feature_data, x='Company', y='Value', hue='Year', ax=axs[i])
    
    axs[i].set_title(feature)
    axs[i].set_ylabel('Value')
    axs[i].set_xlabel('Company')
    axs[i].legend(title='Year')

for j in range(len(features), len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.show()
```
